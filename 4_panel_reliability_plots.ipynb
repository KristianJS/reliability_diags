```python
#dp3
#upper
NAO_unadu_dp3 = np.loadtxt("NAO_"+s+"dp3_coords_upper_tercile.txt")
up_dp3=int(NAO_unadu_dp3.size/4)
NAO_unadju_dp3 = NAO_unadu_dp3.reshape(up_dp3,4)
x_unadu_dp3 = NAO_unadju_dp3[:,0]
y_unadu_dp3 = NAO_unadju_dp3[:,1]
size_unadu_dp3= NAO_unadju_dp3[:,2] 

#lower
NAO_unadl_dp3 = np.loadtxt("NAO_"+s+"dp3_coords_lower_tercile.txt")
low_dp3=int(NAO_unadl_dp3.size/4)
NAO_unadjl_dp3 = NAO_unadl_dp3.reshape(low,4)
x_unadl_dp3 = NAO_unadjl_dp3[:,0]
y_unadl_dp3 = NAO_unadjl_dp3[:,1]
size_unadl_dp3= NAO_unadjl_dp3[:,2]
```

```python
#create reliability plot
#just coordinates no plotting
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

t=np.full((50),0.3333333)
x=np.linspace(0,1,50)
y=np.linspace(0,1,50)
yt=np.linspace(-0.05,1.05,50)

#ASF_20C
#upper tercile data
#adjusted
upper=np.loadtxt("NAO_"+s+"DJF_jet_adj_coords_upper_tercile.txt")
w=int(upper.size/4)
NAO_up = upper.reshape(w,4)
x_u = NAO_up[:,0]
y_u = NAO_up[:,1]
size_u= NAO_up[:,2]*5

p=x_u.reshape(-1,1)
modelu= LinearRegression()
modelu.fit(p,y_u,size_u)
p_new = x_u
q_new = modelu.predict(p_new[:, np.newaxis])
q_reg=modelu.coef_
q_int=modelu.intercept_

uncert_regu =np.loadtxt("unc_reg_u"+s+"DJF_jet_adj.txt")
m1=uncert_regu[0]
m2=uncert_regu[1]
c1=0.3333333-0.3333333*m1
c2=0.3333333-0.3333333*m2

#unadjusted data
p_unu=x_unadu.reshape(-1,1)
q_unu=y_unadu
modelllu= LinearRegression()
modelllu.fit(p_unu,q_unu,size_unadu)
q_unnewu = modelllu.predict(p_unu)
q_regunu=modelllu.coef_
q_intunu=modelllu.intercept_

#uncertainities
uncert_regu_un=np.loadtxt("unc_reg_u"+s+"DJF_unadj.txt")
m1_unu=uncert_regu_un[0]
m2_unu=uncert_regu_un[1]
c1_unu=0.3333333-0.3333333*uncert_regu_un[0]
c2_unu=0.3333333-0.3333333*uncert_regu_un[1]


#lower tercile data
lower=np.loadtxt("NAO_"+s+"DJF_jet_adj_coords_upper_tercile.txt")
wl=int(lower.size/4)
NAO_lo = lower.reshape(wl,4)
x_l = NAO_lo[:,0]
y_l = NAO_lo[:,1]
size_l= NAO_lo[:,2]*5 
col=np.arange(wl)

pl=x_l.reshape(-1,1)
modell= LinearRegression()
modell.fit(pl,y_l,size_l)
p_newl = x_l
q_newl = modell.predict(p_newl[:, np.newaxis])
q_regl=modell.coef_
q_intl=modell.intercept_

uncert_regl =np.loadtxt("unc_reg_l"+s+"DJF_jet_adj.txt")
m1l=uncert_regl[0]
m2l=uncert_regl[1]
c1l=0.3333333-0.3333333*m1l #IMPOSING THAT UNCERTAINTIES GO THROUGH (1/3, 1/3). I THINK THIS MESSES THINGS UP SOMETIMES
c2l=0.3333333-0.3333333*m2l

#plotting unadjusted data
p_unl=x_unadl.reshape(-1,1)
q_unl=y_unadl
modellll= LinearRegression()
modellll.fit(p_unl,q_unl,size_unadl)
p_unnewl = p_unl
q_unnewl = modellll.predict(p_unnewl)
q_regunl=modellll.coef_
q_intunl=modellll.intercept_

#uncertainities unadjusted
uncert_regl_un=np.loadtxt("unc_reg_l"+s+"DJF_unadj.txt")
m1_unl=uncert_regl_un[0]
m2_unl=uncert_regl_un[1]
c1_unl=0.3333333-0.3333333*uncert_regl_un[0]
c2_unl=0.3333333-0.3333333*uncert_regl_un[1]
```

```python
#just coordinates no plotting
#dp3

#upper tercile data
#adjusted
upper_dp3=np.loadtxt("NAO_"+s+"dp3adjusted_coords_upper_tercile.txt")
w_dp3=int(upper_dp3.size/4)
NAO_up_dp3 = upper_dp3.reshape(w_dp3,4)
x_u_dp3 = NAO_up_dp3[:,0]
y_u_dp3 = NAO_up_dp3[:,1]
size_u_dp3= NAO_up_dp3[:,2]*5

p_dp3=x_u_dp3.reshape(-1,1)
modelu_dp3= LinearRegression()
modelu_dp3.fit(p_dp3,y_u_dp3,size_u_dp3)
p_new_dp3 = x_u_dp3
q_new_dp3 = modelu_dp3.predict(p_new_dp3[:, np.newaxis])
q_reg_dp3=modelu_dp3.coef_
q_int_dp3=modelu_dp3.intercept_

uncert_regu_dp3 =np.loadtxt("uncert_reg_u"+s+"dp3adjusted.txt")
m1_dp3=uncert_regu_dp3[0]
m2_dp3=uncert_regu_dp3[1]
c1_dp3=0.3333333-0.3333333*uncert_regu_dp3[0]
c2_dp3=0.3333333-0.3333333*uncert_regu_dp3[1]

#unadjusted data
p_unu_dp3=x_unadu_dp3.reshape(-1,1)
q_unu_dp3=y_unadu_dp3
modelllu_dp3= LinearRegression()
modelllu_dp3.fit(p_unu_dp3,q_unu_dp3,size_unadu_dp3)
q_unnewu_dp3 = modelllu_dp3.predict(p_unu_dp3)
q_regunu_dp3=modelllu_dp3.coef_
q_intunu_dp3=modelllu_dp3.intercept_

#uncertainities
uncert_regu_un_dp3=np.loadtxt("uncert_reg_u"+s+"dp3.txt")
m1_unu_dp3=uncert_regu_un_dp3[0]
m2_unu_dp3=uncert_regu_un_dp3[1]
c1_unu_dp3=0.3333333-0.3333333*uncert_regu_un_dp3[0]
c2_unu_dp3=0.3333333-0.3333333*uncert_regu_un_dp3[1]


#lower tercile data
lower_dp3=np.loadtxt("NAO_"+s+"dp3adjusted_coords_lower_tercile.txt")
wl_dp3=int(lower_dp3.size/4)
NAO_lo_dp3 = lower_dp3.reshape(wl_dp3,4)
x_l_dp3 = NAO_lo_dp3[:,0]
y_l_dp3 = NAO_lo_dp3[:,1]
size_l_dp3= NAO_lo_dp3[:,2]*5 
col_dp3=np.arange(wl_dp3)

pl_dp3=x_l_dp3.reshape(-1,1)
modell_dp3= LinearRegression()
modell_dp3.fit(pl_dp3,y_l_dp3,size_l_dp3)
p_newl_dp3 = x_l_dp3
q_newl_dp3 = modell_dp3.predict(p_newl_dp3[:, np.newaxis])
q_regl_dp3=modell_dp3.coef_
q_intl_dp3=modell_dp3.intercept_

uncert_regl_dp3 =np.loadtxt("uncert_reg_l"+s+"dp3adjusted.txt")
m1l_dp3=uncert_regl_dp3[0]
m2l_dp3=uncert_regl_dp3[1]
c1l_dp3=0.3333333-0.3333333*uncert_regl_dp3[0]
c2l_dp3=0.3333333-0.3333333*uncert_regl_dp3[1]

#plotting unadjusted data
p_unl_dp3=x_unadl_dp3.reshape(-1,1)
q_unl_dp3=y_unadl_dp3
modellll_dp3= LinearRegression()
modellll_dp3.fit(p_unl_dp3,q_unl_dp3,size_unadl_dp3)
p_unnewl_dp3 = p_unl_dp3
q_unnewl_dp3 = modellll_dp3.predict(p_unnewl_dp3)
q_regunl_dp3=modellll_dp3.coef_
q_intunl_dp3=modellll_dp3.intercept_

#uncertainities
uncert_regl_un_dp3=np.loadtxt("uncert_reg_l"+s+"dp3.txt")
m1_unl_dp3=uncert_regl_un_dp3[0]
m2_unl_dp3=uncert_regl_un_dp3[1]
c1_unl_dp3=0.3333333-0.3333333*uncert_regl_un_dp3[0]
c2_unl_dp3=0.3333333-0.3333333*uncert_regl_un_dp3[1]

```
