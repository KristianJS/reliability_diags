```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

s='1980-2010'
```

```python
#ASF_20C
#upper
NAO_unadu = np.loadtxt("NAO_"+s+"DJF_unadj_coords_upper_tercile.txt")
up=int(NAO_unadu.size/4)
NAO_unadju = NAO_unadu.reshape(up,4)
x_unadu = NAO_unadju[:,0]
y_unadu = NAO_unadju[:,1]
size_unadu= NAO_unadju[:,2] 

#lower
NAO_unadl = np.loadtxt("NAO_"+s+"DJF_unadj_coords_lower_tercile.txt")
low=int(NAO_unadl.size/4)
NAO_unadjl = NAO_unadl.reshape(low,4)
x_unadl = NAO_unadjl[:,0]
y_unadl = NAO_unadjl[:,1]
size_unadl= NAO_unadjl[:,2]
```

```python
#dp3
#upper
NAO_unadu_dp3 = np.loadtxt("NAO_"+s+"dp3_coords_upper_tercile.txt")
up_dp3=int(NAO_unadu_dp3.size/4)
NAO_unadju_dp3 = NAO_unadu_dp3.reshape(up_dp3,4)
x_unadu_dp3 = NAO_unadju_dp3[:,0]
y_unadu_dp3 = NAO_unadju_dp3[:,1]
size_unadu_dp3= NAO_unadju_dp3[:,2] 

#lower
NAO_unadl_dp3 = np.loadtxt("NAO_"+s+"dp3_coords_lower_tercile.txt")
low_dp3=int(NAO_unadl_dp3.size/4)
NAO_unadjl_dp3 = NAO_unadl_dp3.reshape(low,4)
x_unadl_dp3 = NAO_unadjl_dp3[:,0]
y_unadl_dp3 = NAO_unadjl_dp3[:,1]
size_unadl_dp3= NAO_unadjl_dp3[:,2]
```

```python
#create reliability plot
#just coordinates no plotting
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

t=np.full((50),0.3333333)
x=np.linspace(0,1,50)
y=np.linspace(0,1,50)
yt=np.linspace(-0.05,1.05,50)

#ASF_20C
#upper tercile data

p_unu=x_unadu.reshape(-1,1)
q_unu=y_unadu
modelllu= LinearRegression()
modelllu.fit(p_unu,q_unu,size_unadu)
q_unnewu = modelllu.predict(p_unu)
q_regunu=modelllu.coef_
q_intunu=modelllu.intercept_

#uncertainities
uncert_regu_un=np.loadtxt("unc_reg_u"+s+"DJF_unadj.txt")
m1_unu=uncert_regu_un[0]
m2_unu=uncert_regu_un[1]
c1_unu=0.3333333-0.3333333*uncert_regu_un[0]
c2_unu=0.3333333-0.3333333*uncert_regu_un[1]


#lower tercile data

#plotting unadjusted data
p_unl=x_unadl.reshape(-1,1)
q_unl=y_unadl
modellll= LinearRegression()
modellll.fit(p_unl,q_unl,size_unadl)
p_unnewl = p_unl
q_unnewl = modellll.predict(p_unnewl)
q_regunl=modellll.coef_
q_intunl=modellll.intercept_

#uncertainities unadjusted
uncert_regl_un=np.loadtxt("unc_reg_l"+s+"DJF_unadj.txt")
m1_unl=uncert_regl_un[0]
m2_unl=uncert_regl_un[1]
c1_unl=0.3333333-0.3333333*uncert_regl_un[0]
c2_unl=0.3333333-0.3333333*uncert_regl_un[1]
```

```python
#just coordinates no plotting
#dp3

#upper tercile data

#unadjusted data
p_unu_dp3=x_unadu_dp3.reshape(-1,1)
q_unu_dp3=y_unadu_dp3
modelllu_dp3= LinearRegression()
modelllu_dp3.fit(p_unu_dp3,q_unu_dp3,size_unadu_dp3)
q_unnewu_dp3 = modelllu_dp3.predict(p_unu_dp3)
q_regunu_dp3=modelllu_dp3.coef_
q_intunu_dp3=modelllu_dp3.intercept_

#uncertainities
uncert_regu_un_dp3=np.loadtxt("uncert_reg_u"+s+"dp3.txt")
m1_unu_dp3=uncert_regu_un_dp3[0]
m2_unu_dp3=uncert_regu_un_dp3[1]
c1_unu_dp3=0.3333333-0.3333333*uncert_regu_un_dp3[0]
c2_unu_dp3=0.3333333-0.3333333*uncert_regu_un_dp3[1]


#lower tercile data

#plotting unadjusted data
p_unl_dp3=x_unadl_dp3.reshape(-1,1)
q_unl_dp3=y_unadl_dp3
modellll_dp3= LinearRegression()
modellll_dp3.fit(p_unl_dp3,q_unl_dp3,size_unadl_dp3)
p_unnewl_dp3 = p_unl_dp3
q_unnewl_dp3 = modellll_dp3.predict(p_unnewl_dp3)
q_regunl_dp3=modellll_dp3.coef_
q_intunl_dp3=modellll_dp3.intercept_

#uncertainities
uncert_regl_un_dp3=np.loadtxt("uncert_reg_l"+s+"dp3.txt")
m1_unl_dp3=uncert_regl_un_dp3[0]
m2_unl_dp3=uncert_regl_un_dp3[1]
c1_unl_dp3=0.3333333-0.3333333*uncert_regl_un_dp3[0]
c2_unl_dp3=0.3333333-0.3333333*uncert_regl_un_dp3[1]
```

```python
#=====================================================================
#
#      KJS EDITED PLOT
#
#=====================================================================

#create reliability plot
#just plotting
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

fig, axs = plt.subplots(2,2, figsize=(7,7), dpi=500, constrained_layout=True)

#plt.figure(figsize=(5, 5))
fig.suptitle('NAO terciles of DJF '+s)

for i in range(2):
    for j in range(2):
        axs[i,j].set_aspect(1)
        axs[i,j].plot(x,t,c='black',linewidth=0.5)
        axs[i,j].plot(t,yt,c='black', linewidth=0.5)
        axs[i,j].plot(x,y,c='black')
        axs[i,j].set_xlim([0,1])
        axs[i,j].set_ylim([-0.05, 1.05])
        axs[i,j].set_xlabel('forecast probability')
        axs[i,j].set_ylabel('observed frequency')

        
#ASF_20C
#upper tercile data
axs[0,0].scatter(x_unadu, y_unadu, c='r', alpha=0.5, s=size_unadu*5)
axs[0,0].scatter(x_unadu, y_unadu, c='black', s=0.5)
axs[0,0].plot(x,q_regunu*x + q_intunu, c='red')
axs[0,0].fill_between(x, m1_unu*x+c1_unu, m2_unu*x+c2_unu,alpha=0.05,color='red')

#lower tercile data
axs[0,1].scatter(x_unadl, y_unadl, c='r',alpha=0.5, s=size_unadl*5)
axs[0,1].scatter(x_unadl, y_unadl, c='black', s=0.5)
axs[0,1].plot(x,q_regunl*x + q_intunl, c='red')
axs[0,1].fill_between(x, m1_unl*x+c1_unl, m2_unl*x+c2_unl,alpha=0.05,color='red')


axs[0,0].set_title('a) upper tercile ECMWF')
axs[0,1].set_title('b) lower tercile ECMWF')

x_dp3=x
#dp3
#upper tercile data
print(x_unadu_dp3)
print(y_unadu_dp3)
print(size_unadu_dp3*5)
axs[1,0].scatter(x_unadu_dp3, y_unadu_dp3, c='r', alpha=0.5, s=size_unadu_dp3*5)
axs[1,0].scatter(x_unadu_dp3, y_unadu_dp3, c='black', s=0.5)
axs[1,0].plot(x_dp3,q_regunu_dp3*x_dp3 + q_intunu_dp3, c='red')
axs[1,0].fill_between(x_dp3, m1_unu_dp3*x_dp3+c1_unu_dp3, m2_unu_dp3*x_dp3+c2_unu_dp3,alpha=0.05,color='red')

#lower tercile data
axs[1,1].scatter(x_unadl_dp3, y_unadl_dp3, c='r',alpha=0.5, s=size_unadl_dp3*5)
axs[1,1].scatter(x_unadl_dp3, y_unadl_dp3, c='black', s=0.5)
axs[1,1].plot(x_dp3,q_regunl_dp3*x_dp3 + q_intunl_dp3, c='red')
axs[1,1].fill_between(x_dp3, m1_unl_dp3*x_dp3+c1_unl_dp3, m2_unl_dp3*x_dp3+c2_unl_dp3,alpha=0.05,color='red')

axs[0,0].annotate('RPC=1.8',(0.05,0.97), fontsize=8.5)
axs[1,0].annotate('RPC=2.3',(0.05,0.97), fontsize=8.5)

axs[1,0].set_title('c) upper tercile UKMO')
axs[1,1].set_title('d) lower tercile UKMO')
fig.set_constrained_layout_pads( w_pad=1 / 100, h_pad=10/ 100, hspace=0, wspace=0)

#print(x_unadu, size_unadu, size_unadu.sum())


plt.show()
```
